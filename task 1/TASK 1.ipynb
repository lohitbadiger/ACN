```python
# YOU MUST WRITE A PYTHON SELENIUM SCRIPT TO DRAG ROWS IN A PARTICULAR GIVEN ORDER FOR HTML PAGE



# import required libreries


from selenium import webdriver

# ActionChains are a way to automate low level interactions such as mouse movements, 
# mouse button actions, key press, and context menu interactions. 
# This is useful for doing more complex actions like hover over and drag and drop.
from selenium.webdriver.common.action_chains import ActionChains


# WebDriverWait is applied on certain element with defined expected condition and time. This wait is only applied to the specified element. This wait can also throw exception when element is not found
from selenium.webdriver.support.ui import WebDriverWait   





# checking an element is visible and enabled such that we can click on the element. We need to pass wait time and the locator as parameters.
from selenium.webdriver.support import expected_conditions as EC



# Set of supported locator strategies
from selenium.webdriver.common.by import By


# import timing => wait until either some event happens or maximum time reached.
import time
driver = webdriver.Chrome(executable_path=r'D:\\softwares\\chromedriver.exe')

driver.get('D:\\task\\ACN InterviewTask-master\\InterviewTask-master\\TASK 1\\index.html')

# using sleep() to hault the code execution 
time.sleep(2)

# in this order i want the result

jab_code_order =['J92.9','A10.2','E24.9','B10.2','F19.10','D10.11']


print('----------------------------------------------------------------------------------')
# creating function get the current order of the items

def get_current_element_order():
#  finding the elements from table throw ,method called  'find_elements_by_xpath', which will match the elements by xpath
    array_of_elements = driver.find_elements_by_xpath("//tbody//tr")
    return array_of_elements

print('----------------------------------------------------------------------------------')


# creating the functions to identify the locations of the elements 

def map_text_location():
#     first we will create an empty list called text_array_to_return , to which we will add the elements later
    text_array_to_return = []
#     here we looping through all the elements using for loop, we will iterate one by one as 'element'
    for element in get_current_element_order():
#         now we are extracting the elements text and storing the text in the list which we called above "text_array_to_return"
        text_array_to_return.append(element.text)
#     finally we will return the text_array_to_return which is now filled with all of the texts.
    return text_array_to_return


print('----------------------------------------------------------------------------------')


# now we want to get  the button orders 
# so for that we created below function 

# 1. first we will create an empty dict
# 2. from the above 'map_text_location' fun we will take length 
# 3. so that we can iterate to that length of iteraions

def get_current_button_order():
    # creating the empty dictionary below called "get_current_button_order_dict"
    
    get_current_button_order_dict = {}
# nested for loop
    for value1 in range(len(map_text_location())):
        
        for value2 in get_current_element_order():
            # find_element_by_css_selector => Finds an element by css selector and store them variable 'a'
            # here we are using format method 
            a = driver.find_element_by_css_selector("tbody tr:nth-child({})".format(value1 + 1))
            
            # checking wheather value2 is equal to the text content of 'a' variable 
            
            if value2.text == a.text:
                
            # if True 
            # nth child row will be affected 
            # The :nth-child() CSS pseudo-class matches elements based on their position in a group of siblings.
            # * get_current_button_order_dict have index as value2.text 
            
            # fills the dict will the current position of the elements
            
                get_current_button_order_dict[value2.text] = driver.find_element_by_css_selector("tbody tr:nth-child({}) td:nth-child(1)".format(value1+ 1))
                
                
    return get_current_button_order_dict

# process to assing the current values order to required place
# iterate the items present in jab_code_order

for item in range(len(jab_code_order)):
    # assign the 'get_current_button_order' values to element_dict
    element_dict = get_current_button_order()
    # taken empty string as value for this_key variable
    this_key = ""
    # iterate all key values from the element_dict which we just now assigned above
    for key in element_dict:
        
        if jab_code_order[item] in key:
            this_key = key
    source_element = element_dict[this_key]
    current_order_of_elements = get_current_element_order()
    dest_element = current_order_of_elements[item]

    if dest_element.location['y'] - source_element.location['y'] < 0:
        print(source_element,0,dest_element.location['y'] - source_element.location['y'] - 1)
        
        
        
    #  ActionChains(driver).drag_and_drop_by_offset
# A convenience method that performs click-and-hold at the location of the source element, moves by a given offset, then releases the mouse.

        ActionChains(driver).drag_and_drop_by_offset(source_element,
                                                     0,
                                                     dest_element.location['y'] - source_element.location['y'] - 5).perform()
    else:
        print("else")
        ActionChains(driver).drag_and_drop_by_offset(source_element,
                                                     0,
                                                     dest_element.location['y'] - source_element.location['y']).perform()
        
     
    #  WebDriverWait
        # This waits up to 5 seconds before throwing a TimeoutException unless it finds the element to return within 5 seconds.
        
        # Expects an element identified by the locator to be present in the web page and expects the element to be in selected state
    WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.CSS_SELECTOR,"tbody tr:nth-child(1) td:nth-child(1)")))
```
